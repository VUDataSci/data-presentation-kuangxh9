```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data & Source:** <br>
`national-gdp-world.csv` [Source Link](https://datacatalog.worldbank.org/search/dataset/0037712/World-Development-Indicators)

```python
nation_gdp = pd.read_csv('national-gdp-world.csv')
```

```python
selected_countries = nation_gdp[nation_gdp['Year'] == 2021].sort_values(by='GDP (constant 2015 US$)', ascending=False).dropna()['Entity'][1:6]
```

```python
list(selected_countries.values)
```

```python
selected_nation_gdp = nation_gdp[nation_gdp['Entity'].isin(selected_countries)]
pivoted_selected_nation_gdp = selected_nation_gdp.pivot(index='Year', columns='Entity', values='GDP (constant 2015 US$)')

sns.set(style="ticks")
plt.figure(figsize=(14, 8))
sns.lineplot(data=pivoted_selected_nation_gdp, linewidth=3)
plt.grid(linewidth=1, linestyle=':')
plt.title('GDP Change Over Time for Selected Entities (All Years)', fontsize='24', pad=12.5)
plt.xlabel('Year', fontsize='22', labelpad =12.5)
plt.ylabel('GDP', fontsize='22', labelpad =12.5)
plt.legend(title='Country', title_fontsize='16', fontsize='16', loc='upper left')
plt.grid(visible=True)
plt.show()
```
